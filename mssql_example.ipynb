```python
# example showing read from MSSQL

from pyspark import SparkContext
from pyspark.sql import SQLContext, HiveContext, SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession \
        .builder \
        .appName("mssql_test") \
        .master("locatl[*]") \
        .config("spark.jars", "file:///path/to/jars/mssql-jdbc-8.4.1.jre8.jar") \
        .config("spark.executor.extraClassPath", "file:///path/to/jars/mssql-jdbc-8.4.1.jre8.jar") \
        .config("spark.executor.extraLibrary", "file:///path/to/jars/mssql-jdbc-8.4.1.jre8.jar") \
        .config("spark.driver.extraClassPath", "file:///path/to/jars/mssql-jdbc-8.4.1.jre8.jar") \
        .enableHiveSupport() \
        .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

database = 'database1'
table = 'dbo.table1'
user = ''
pwd = ''
url = "jdbc:sqlseerver://servername\instance:port;databaseName=" + database + ";"

jdbcDF = spark.read \
            .format("jdbc") \
            .option("driver", "com.microsoft.sqlserver.jdbc.SQLServerDriver") \
            .option("url", url) \
            .option("dbtable", table) \
            .option("user", user) \
            .option("password", pwd) \
            .load()

jdbcDF.show(10)
```
